```python
from datasets import load_dataset

ds_sun = load_dataset("cointegrated/panlex-meanings", name="sun", split="train")
ds_ind = load_dataset("cointegrated/panlex-meanings", name="ind", split="train")
df_sun = ds_sun.to_pandas()
df_ind = ds_ind.to_pandas()

df_ind_sun = df_sun.merge(df_ind, on="meaning", suffixes=["_sun", "_ind"]).drop_duplicates(
    subset=["txt_sun", "txt_ind"]
)
df_ind_sun = df_ind_sun[["txt_ind", "txt_sun"]]
df_ind_sun
```

```python
import pandas as pd

df_nusax = pd.read_csv("https://raw.githubusercontent.com/IndoNLP/nusax/refs/heads/main/datasets/lexicon/sundanese.csv")
df_nusax = df_nusax[["indonesian", "sundanese"]]
df_nusax.columns = ["txt_ind", "txt_sun"]
df_nusax
```

```python
df = pd.concat([df_nusax, df_ind_sun], ignore_index=True).drop_duplicates(subset=["txt_ind"], keep="first")
df
```

```python
df.to_csv("../data/ind_sun_lexicon.csv", index=False)
```
