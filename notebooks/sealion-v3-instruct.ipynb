```python
import pandas as pd

df = pd.read_csv("../public_data/dev/track_a/sun_a.csv")
labels = ["Anger", "Disgust", "Fear", "Joy", "Sadness", "Surprise"]
df.loc[:, labels] = ""
df
```

```python
from ollama import Client

client = Client(host="http://localhost:11434")
```

```python
from tqdm.auto import tqdm

emotions = ["marah", "jijik", "takut", "senang", "sedih", "terkejut"]
prompt = "Apakah teks berikut mengekspresikan emosi {emotion}? Jawab ya atau tidak.\nKalimat: {text}\nJawaban: "

texts = df["text"].tolist()

for idx, text in enumerate(tqdm(texts)):
    predictions = []
    for emotion in emotions:
        response = client.chat(
            model="hf.co/aisingapore/gemma2-9b-cpt-sea-lionv3-instruct-gguf:Q4_K_M",
            messages=[
                {
                    "role": "user",
                    "content": prompt.format(emotion=emotion, text=text),
                },
            ],
        )
        prediction = response["message"]["content"].lower()
        predictions.append(prediction)
    df.loc[idx, labels] = predictions
```

```python
df
```

```python
for label in labels:
    df[label] = df[label].apply(lambda x: x.split("\n")[0].replace(".", "").strip())
```

```python
for label in labels:
    df[label] = df[label].map({"ya": 1, "tidak": 0})
```

```python
df.drop("text", axis=1).to_csv("sealion-v3-instruct.csv", index=False)
```
